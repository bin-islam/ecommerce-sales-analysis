```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/data.csv', encoding='ISO-8859-1')
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
df.head()
```

```python
# Drop nulls
df = df.dropna(subset=['CustomerID', 'Description'])

# Remove cancellations (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove bad quantity/price rows
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Fix dtypes
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype(int)

# Engineer revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']

print(df.shape)
df.head()
```

```python
print("Date range:", df['InvoiceDate'].min(), "to", df['InvoiceDate'].max())
print("Unique customers:", df['CustomerID'].nunique())
print("Unique products:", df['StockCode'].nunique())
print("Countries:", df['Country'].nunique())
print("Total revenue: £{:,.2f}".format(df['Revenue'].sum()))

# Save cleaned data for use in analysis
df.to_csv('../data/data_cleaned.csv', index=False)
print(f"\nCleaned data saved to ../data/data_cleaned.csv ({df.shape[0]} rows)")
```
